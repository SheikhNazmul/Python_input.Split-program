### Generator 


```python
lst = [ x for x in range (500)] # dataset 

def deta_loder(chunk_size,lst):
        for i in range(len(lst)):
            yield lst[i:i+chunk_size]


x = deta_loder (5,lst)

print (next(x))
print (next(x))
print (next(x))
print (next(x))
print (next(x))
print (x)
print (type(x))
```

```python
lst = [ x for x in range (500)] # dataset 

def deta_loder(chunk_size,lst):
        for i in range(0,len(lst),chunk_size):
            yield lst[i:i+chunk_size]


x = deta_loder (5,lst)

print (next(x))
print (next(x))
print (next(x))
print (next(x))
print (next(x))
print (x)
print (type(x))
```

### Map finction 

```python
lst = [1,2,3,4,5,6]

def squere (x):
    return x**2

lst = list(map(squere ,lst))

print (lst)
```

```python
## string 

string = "Hello World , Welcome to the age of AI"

string1 = list(map(str.upper,string))

print (string1)

string1 = " ".join(string1)
print (string1)
```

```python
## map with lambda function

lst = [1,2,3,4,5,6]
lst = list(map(lambda x: x**2 if x%2==0 else x ,lst))

print(lst)
```

### filter 

```python
numbers = [x for x in range (100)]


#even
even = filter(lambda x : x % 2==0,numbers)
print (list(even))

#fifty upper
fifty_upper = filter(lambda x : x >50 ,numbers)
print (list(fifty_upper))
```

```python
data = [0,1,'',None, "hello", [],[1,2],None,True,False,None ]

cleaned_data = list(filter(None,data))
print (cleaned_data)
```
